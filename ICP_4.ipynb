```python
import pandas as pd
```

```python
df = pd.read_csv("former_names.csv")
```

```python
print(df.head())
print(df.tail())
```

```python
print(df.info())
print(df.describe())
```

```python
print(df[['current', 'former']])
```

```python
print(df[df['current'] == 'DR Congo'])
```

```python
print(df.loc[10:20, ['current', 'former']])
print(df.iloc[10:20, 0:3])
```

```python
print(df.isnull().sum())
```

```python
df['current'].fillna(df['current'].mode(), inplace=True)
```

```python
df.dropna(axis=0)
df.dropna(axis=1)
```

```python
df.drop_duplicates(inplace=True)
```

```python
df['start_date'] = pd.to_datetime(df['start_date'], errors='ignore')
```

```python
df.columns = df.columns.str.lower().str.replace(" ", "_")
```

```python
df
```

```python
df.groupby('current')['start_date'].mean()
```

```python
df.sort_values(by=['start_date','end_date'], ascending=[True,False])
```

```python
df.set_index('current', inplace=True)
df.reset_index(inplace=True)
```

```python
df1 = df[['current','start_date']]
df2 = df[['current','end_date']]

pd.merge(df1, df2, on='current', how='inner')
pd.merge(df1, df2, on='current', how='outer')
pd.merge(df1, df2, on='current', how='left')
pd.merge(df1, df2, on='current', how='right')
```

```python
pd.concat([df1, df2], axis=0)
pd.concat([df1, df2], axis=1)
```

```python
import matplotlib.pyplot as plt

df['end_date']   = pd.to_datetime(df['end_date'])
df['duration_years'] = (df['end_date'] - df['start_date']).dt.days / 365

plt.figure(figsize=(10,5))
plt.barh(df['current'], df['duration_years'], color='skyblue')
plt.xlabel("Years under former name")
plt.ylabel("Country")
plt.title("Duration of Former Country Names")
plt.gca().invert_yaxis()
plt.show()
```
