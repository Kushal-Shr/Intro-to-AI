a. Using NumPy create random vector of size 15 having only Integers in the range 1-20.
1. Reshape the array to 3 by 5
2. Print array shape.
3. Replace the max in each row by 0

```python
import numpy as np
```

```python
arr = np.random.randint(1, 21, 15)
print(arr)
```

```python
reshaped_arr = arr.reshape(3, 5)
print(reshaped_arr)
```

```python
print(reshaped_arr.shape)
```

```python
for i in range(reshaped_arr.shape[0]):
    max_index = np.argmax(reshaped_arr[i])
    reshaped_arr[i][max_index] = 0

print(reshaped_arr)
```

b. Create a 2-dimensional array of size 4 x 3 (composed of 4-byte integer elements), also print the shape, type and data
type of the array.

```python
arr = np.zeros((4, 3), dtype=np.int32)
print(arr)

print(arr.shape)
print(type(arr))
print(arr.dtype)
```

c. Write a program to compute the eigenvalues and right eigenvectors of a given square array given below:
[[ 3 -2]
[ 1 0]]

```python
A = np.array([[3, -2],
              [1,  0]])

eigenvalues, eigenvectors = np.linalg.eig(A)

print("Eigenvalues: ", eigenvalues,)
print("Right eigenvectors:\n", eigenvectors)
```

d. Compute the sum of the diagonal element of a given array.
[[0 1 2]
[3 4 5]]

```python
arr = np.array([[0, 1, 2],
                [3, 4, 5]])

diag_sum = np.trace(arr)
print(diag_sum)
```

e. Write a NumPy program to create a new shape to an array without changing its data.
Reshape 3x2:
[[1 2]
[3 4]
[5 6]]
Reshape 2x3:
[[1 2 3]
[4 5 6]]


```python
arr = np.array([[1, 2],
                [3, 4],
                [5, 6]])

reshaped = arr.reshape(2, 3)
print(arr)
print(reshaped)
```
