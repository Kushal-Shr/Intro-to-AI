```python
grade = int(input("Enter your grade: "))

if grade >= 92:
    print("Your grade is A")
elif 85 <= grade < 92:
    print("Your grade is B")
elif 80 <= grade < 85:
    print("Your grade is C")
elif 70 <= grade < 80:
    print("Your grade is D")
elif 60 <= grade < 70:
    print("Your grade is E")
else:
    print("You failed buddy")
```
