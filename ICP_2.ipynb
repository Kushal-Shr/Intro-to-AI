1. Display this pattern:
```
*
* *
* * *
* * * *
* * * * *
* * * *
* * *
* *
*
```

````python
for i in range(1, 6):
    j = 1

    while j <= i:
        print('*', end = ' ')
        j += 1

    print(end = '\n')

for i in range(1, 5):
    j = 4

    while j >= i:
        print('*', end = ' ')
        j -= 1

    print(end = '\n')
````

````python
def sum_all(args):
    if not args:
        return 0
    return sum_all(args[1:]) + args[0]

print(f'Sum of 1, 2, 3 is: {sum_all([1, 2, 3])}')
print(f'Sum of 1, 2, 3, 4, 5 is: {sum_all([1, 2, 3, 4, 5])}')
````

````python
class Employee:
    count = 0

    def __init__(self, name, family, salary, department):
        self.name = name
        self.family = family
        self.salary = salary
        self.department = department
        Employee.count += 1

    def avg_salary(self, employees):
        sum_salary = 0
        for employee in employees:
            sum_salary += employee.salary
        if Employee.count > 0:
            return int(sum_salary / Employee.count)
        else:
            return 0

class FulltimeEmployee(Employee):
    pass

fulltime_employee1 = Employee("Kushal", "Family A", 220000, "Data Services")
employee2 = Employee("Roshit", "Family C", 160000, "Data Science")

fulltime_employee3 = FulltimeEmployee("Muaz", "Family B", 200000, "Web Dev")
employee4 = Employee("Tanguy", "Family D", 170000, "Cloud")

all_employees = [fulltime_employee1, employee2, fulltime_employee3, employee4]

average_salary_employee = employee2.avg_salary(all_employees)
print(f"Average salary calculated by Employee instance: {average_salary_employee}")

average_salary_fulltime = fulltime_employee1.avg_salary(all_employees)
print(f"Average salary calculated by FulltimeEmployee instance: {average_salary_fulltime}")

print(f"Total number of employees: {Employee.count}")
````
